```python
import nltk
import pandas as pd, numpy as np
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import CountVectorizer
import spacy
```

```python
df0 = pd.read_csv('data 1_train.csv', na_values=['#NAME?'])
df1 = pd.read_csv('data 2_train.csv', na_values=['#NAME?'])
df0.rename(columns={' text':'text', ' aspect_term':'aspect_term' , ' term_location':'term_location', ' class':'class1'},inplace=True)
df1.rename(columns={' text':'text', ' aspect_term':'aspect_term' , ' term_location':'term_location', ' class':'class1'},inplace=True)
```

```python
df0['text'] = df0['text'].apply(lambda x: x.replace('[comma]',','))
df0['text'] = df0['text'].apply(lambda x: x.lower())
df0['text'] = df0['text'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))

operators=set(('and','but', 'however','although','even though','yet','nevertheless','in spite of','by means of','in place of','in case of',
              'on behalf of','with respect to','in accordance with','in front of','on top of','in addition to', 'in lieu of','on account of',
              'with regard to','like'))
stop = set(stopwords.words('english'))-operators
df0['text_new'] = df0['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))


# print(df0.head())
# print(df0['text_new'])
y0 = df0['class1'].values
X0 = df0['text_new']

df1['text'] = df1['text'].apply(lambda x: x.replace('[comma]',','))
df1['text'] = df1['text'].apply(lambda x: x.lower())
df1['text'] = df1['text'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))


df1['text_new'] = df1['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
# print(df1.head())
# print(df1['text_new'])
```

```python
df0['features1']=""
df0['features2']=""
df0['features3']=""
df0['features4']=""
df0['features5']=""
df0['features6']=""

df1['features11']=""
df1['features12']=""
df1['features13']=""
df1['features14']=""
df1['features15']=""
df1['features16']=""
```

```python
#subject-noun rule
from spacy.symbols import nsubj, VERB
nlp=spacy.load('en')
#head is verb and subject is child
# Finding a verb with a subject from below â€” good
#verbs = {}
for num,text in enumerate(df0['text_new']):
        text = nlp(text)
        for possible_subject in text:
            if possible_subject.dep == nsubj and possible_subject.head.pos == VERB:
                new_feat=str(possible_subject)+" "+str(possible_subject.head)
                df0['features1'].iloc[num]=new_feat
                #verbs[possible_subject]=possible_subject.head
                
for num,text in enumerate(df1['text_new']):
        text = nlp(text)
        for possible_subject in text:
            if possible_subject.dep == nsubj and possible_subject.head.pos == VERB:
                new_feat=str(possible_subject)+" "+str(possible_subject.head)
                df1['features11'].iloc[num]=new_feat
```

```python
X0 = df0[['text_new','features1','features2','features5', 'features6','aspect_term']]
X0dict = X0.to_dict('record')
# print(X0dict)

X1 = df1[['text_new','features11','features12','features15', 'features16','aspect_term']]
X1dict = X1.to_dict('record')
# print(X1dict)

y0 = df0['class1']
y1 = df1['class1']
```

```python
from sklearn.feature_extraction import DictVectorizer
v = DictVectorizer(sparse=False)
X0 = v.fit_transform(X0dict)
# print(X0)

X1 = v.fit_transform(X1dict)
# print(X1)
```

```python
from sklearn.svm import LinearSVC
lsvc = LinearSVC(random_state = 0, C=0.1)
svm1 = lsvc

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score

# print(X0)
skf = StratifiedKFold(n_splits=10)
skf.get_n_splits(X0, y0)
  
sum1 = 0
sp = 0
sr = 0
sf1 = 0
cms = 0 #find avg of cmatrix
print('Class 1 -> label -1, Class 2 -> label 0, Class 3 -> label 1')
print('\n')
print('>>> Results from 10 - fold Cross Validation <<<')
# print('\n')
for train_index, test_index in skf.split(X0, y0):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X0[train_index], X0[test_index]
    y_train, y_test = y0[train_index], y0[test_index]
    
    svm1.fit(X_train, y_train)
    preds = svm1.predict(X_test)
    np.savetxt("output1.txt",preds, newline="\n", fmt="%d")

    cm = (confusion_matrix(y_test, preds.round()))
    cms = cms + cm

    acc = (accuracy_score(y_test, preds.round()))
    p = (precision_score(y_test, preds.round(), average='weighted'))
    r = (recall_score(y_test, preds.round(), average='weighted'))
    f1 = (f1_score(y_test, preds.round(), average='weighted'))
    sp = sp + p
    sr = sr + r
    sf1 = sf1 + f1
    sum1 = sum1 + acc

from sklearn.externals import joblib
joblib.dump(svm1, 'svm1.pkl') 

cms = cms/10

for i in range(0,3):
    p = cms[i][i]/(cms[i][0]+cms[i][1]+cms[i][2])
    r = cms[i][i]/(cms[0][i]+cms[1][i]+cms[2][i])
    f1 = 2*p*r/(p+r)
    print('Precision for class ' + str(i+1) + ': ' + str(p))
    
    print('Recall for class ' + str(i+1) + ': ' + str(r))
    
    print('F1 Score for class ' + str(i+1) + ': ' + str(f1))
    print('\n')

avg_acc = sum1/10
avg_p = sp/10
avg_r = sr/10
avg_f1 = sf1/10

print('Overall Average Accuracy = ' + str(avg_acc))
print('\n')
```

```python
from sklearn.svm import LinearSVC
lsvc = LinearSVC(random_state = 0, C=0.1)
svm2 = lsvc

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score

skf = StratifiedKFold(n_splits=10)
skf.get_n_splits(X1, y1)
  
sum1 = 0
sp = 0
sr = 0
sf1 = 0
cms = 0 #find avg of cmatrix
print('Class 1 -> label -1, Class 2 -> label 0, Class 3 -> label 1')
print('\n')
print('>>> Results from 10 - fold Cross Validation <<<')

for train_index, test_index in skf.split(X1, y1):
    X_train, X_test = X1[train_index], X1[test_index]
    y_train, y_test = y1[train_index], y1[test_index]

    svm2.fit(X_train, y_train)
    preds = svm2.predict(X_test)
    np.savetxt("output2.txt",preds, newline="\n", fmt="%d")

    cm = (confusion_matrix(y_test, preds.round()))

    cms = cms + cm

    acc = (accuracy_score(y_test, preds.round()))
    p = (precision_score(y_test, preds.round(), average='weighted'))
    r = (recall_score(y_test, preds.round(), average='weighted'))
    f1 = (f1_score(y_test, preds.round(), average='weighted'))
    sp = sp + p
    sr = sr + r
    sf1 = sf1 + f1
    sum1 = sum1 + acc

from sklearn.externals import joblib
joblib.dump(svm2, 'svm2.pkl') 
    
cms = cms/10

for i in range(0,3):
    p = cms[i][i]/(cms[i][0]+cms[i][1]+cms[i][2])
    r = cms[i][i]/(cms[0][i]+cms[1][i]+cms[2][i])
    f1 = 2*p*r/(p+r)
    print('Precision for class ' + str(i+1) + ': ' + str(p))
    
    print('Recall for class ' + str(i+1) + ': ' + str(r))
    
    print('F1 Score for class ' + str(i+1) + ': ' + str(f1))
    print('\n')

avg_acc = sum1/10
avg_p = sp/10
avg_r = sr/10
avg_f1 = sf1/10

print('Overall Average Accuracy = ' + str(avg_acc))
print('\n')
```
