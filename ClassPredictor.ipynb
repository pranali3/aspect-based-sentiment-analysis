```python
import nltk
import pandas as pd, numpy as np
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import CountVectorizer
import spacy
```

```python
df0 = pd.read_csv('data 1_.csv', na_values=['#NAME?'])
df1 = pd.read_csv('data 2_.csv', na_values=['#NAME?'])
df0.rename(columns={' text':'text', ' aspect_term':'aspect_term' , ' term_location':'term_location', ' class':'class1'},inplace=True)
df1.rename(columns={' text':'text', ' aspect_term':'aspect_term' , ' term_location':'term_location', ' class':'class1'},inplace=True)
```

```python
df0['text'] = df0['text'].apply(lambda x: x.replace('[comma]',','))
df0['text'] = df0['text'].apply(lambda x: x.lower())
df0['text'] = df0['text'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))

operators=set(('and','but', 'however','although','even though','yet','nevertheless','in spite of','by means of','in place of','in case of',
              'on behalf of','with respect to','in accordance with','in front of','on top of','in addition to', 'in lieu of','on account of',
              'with regard to','like'))
stop = set(stopwords.words('english'))-operators
df0['text_new'] = df0['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))


# print(df0.head())
# print(df0['text_new'])
y0 = df0['class1'].values
X0 = df0['text_new']

df1['text'] = df1['text'].apply(lambda x: x.replace('[comma]',','))
df1['text'] = df1['text'].apply(lambda x: x.lower())
df1['text'] = df1['text'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))


df1['text_new'] = df1['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
# print(df1.head())
# print(df1['text_new'])
```

```python
df0['features1']=""
df0['features2']=""
df0['features3']=""
df0['features4']=""
df0['features5']=""
df0['features6']=""

df1['features11']=""
df1['features12']=""
df1['features13']=""
df1['features14']=""
df1['features15']=""
df1['features16']=""
```

```python
#subject-noun rule
from spacy.symbols import nsubj, VERB
nlp=spacy.load('en')
#head is verb and subject is child
# Finding a verb with a subject from below â€” good
#verbs = {}
for num,text in enumerate(df0['text_new']):
        text = nlp(text)
        for possible_subject in text:
            if possible_subject.dep == nsubj and possible_subject.head.pos == VERB:
                new_feat=str(possible_subject)+" "+str(possible_subject.head)
                df0['features1'].iloc[num]=new_feat
                #verbs[possible_subject]=possible_subject.head
                
for num,text in enumerate(df1['text_new']):
        text = nlp(text)
        for possible_subject in text:
            if possible_subject.dep == nsubj and possible_subject.head.pos == VERB:
                new_feat=str(possible_subject)+" "+str(possible_subject.head)
                df1['features11'].iloc[num]=new_feat
```
